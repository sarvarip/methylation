```python
import numpy as np
filelist = []
with open('filenames.txt') as f:
    for line in f:
        filelist.append(line.strip())
```

```python
command_list0 = []
for j in range(0,len(filelist),1):
    file_ = filelist[j]
    file_in = file_
    file_out = file_.replace('.fastq', '_cutadapt.fastq')
    command_list0.append('cutadapt -a CTGTCTCTTATA -o ' + file_out + ' ' + file_in)
```

```python
with open('command_list0.txt', 'w+') as cl:
    for command in command_list0:
        cl.write(command + '\n')
```
